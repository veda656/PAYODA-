```python
print(df.isnull().sum())
#this shows the number of null values, i.e, the the columns which do not have values. 
#In steroid we have 1 missing data, we have 67 missing data in protime. 
#So it will be good if we ignore protime when analysing the data since it is having a lot of missing values considered to others.
```

```python
a= ['age','albumin','sgot','alk_phosphate','bilirubin']
for i in range(0,len(a)):
    sns.distplot(df[a[i]], color='b', bins=100, hist_kws={'alpha':0.4})
    plt.show()
#this graphs show us where the data is centred, for age the data is centred at 40.
```

```python
v = ['age','albumin','sgot','alk_phosphate','bilirubin','protime']
for i in range(0,len(v)):
    sns.boxplot(df[v[i]])
    plt.show()
#this plot detects outliers. Considering the age, 80 is an outlier.
```

```python
#heatmap for the data which show us the correlation between attributes.
c=df.corr()
sns.heatmap(c,cmap="BrBG", annot=True)
c
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows:", duplicate_rows_df.shape)
#this shows that there are no duplicate rows or columns.
```

```python
#removing protime since it has many missing values.
dff=df.drop(['protime'],axis=1)
```

```python
dff=dff.dropna()
print(dff.isnull().sum())
dff.count()
```

```python
#heatmap after removing protime.
c=dff.corr()
sns.heatmap(c,cmap="BrBG", annot=True)
c
```
